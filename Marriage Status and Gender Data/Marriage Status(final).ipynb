```python
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'poverty_female': [765.54, 736.53, 709.71, 689.76, 685.09],
}

average_income = pd.DataFrame(data)
average_income = average_income.set_index('Year')

# Scatter plot
plt.scatter(average_income.index, average_income['poverty_female'], label='poverty_female', color='hotpink')

# Calculate the linear regression
slope, intercept, r_value, p_value, std_err = linregress(average_income.index, average_income['poverty_female'])

# Plot the regression line
regression_line = slope * average_income.index + intercept
plt.plot(average_income.index, regression_line, label='Regression Line', color='blue')

# Customize the plot
plt.title('Average Female Income with Regression Line')
plt.xlabel('Year')
plt.ylabel('Poverty Level')
plt.legend()
plt.grid(True)

# Show the plot
print(f'R-value: {r_value}')


plt.xticks(average_income.index)
plt.show()
```

```python

data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'poverty_female': [765.54, 736.53, 709.71, 689.76, 685.09],
    'poverty_male': [616.85, 590.72, 567.75, 555.62, 554.38],
    'poverty_family': [249.85, 237.54, 228.49, 221.80, 220.88]
}

average_income_family = pd.DataFrame(data)
average_income_family = average_income_family.set_index('Year')

# Plotting scatter plots and regression lines for each variable
fig, ax = plt.subplots()
colors = {'poverty_female': 'hotpink', 'poverty_male': 'blue', 'poverty_family': 'orange'}


for column in average_income_family.columns:
    # Scatter plot
    ax.scatter(average_income_family.index, average_income_family[column], label=column, color=colors[column])

    # Calculate the linear regression
    slope, intercept, r_value, p_value, std_err = linregress(average_income_family.index, average_income_family[column])

    # Plot the regression line
    regression_line = slope * average_income_family.index + intercept
    ax.plot(average_income_family.index, regression_line, label=f'Regression Line ({column})', color=colors[column])

# Customize the plot
ax.set_title('Income based on Marriage Status')
ax.set_xlabel('Year')
ax.set_ylabel('Poverty Level')
ax.grid(True)
ax.set_xticks(average_income_family.index)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))


# Show the plot
print(f'R-value: {r_value}')

plt.show()

```

```python



# Extracting the relevant columns
data_female = selected_data['poverty_female']
data_male = selected_data['poverty_male']
data_family = selected_data['poverty_family']

# Setting the significance level (alpha)
alpha = 0.05  # Change this to your desired significance level

# Performing one-way ANOVA
anova_result = f_oneway(data_female, data_male, data_family)

# Displaying ANOVA results
print("ANOVA Results:")
print(anova_result)

# Checking if the p-value is less than the significance level
if anova_result.pvalue < alpha:
    print("Reject the null hypothesis - There are significant differences among groups.")
else:
    print("Fail to reject the null hypothesis - No significant differences among groups.")
```
